This is an attempt to catalogue each player's rookie year performances

```python
import pandas as pd
```

```python
import sqlite3
```

```python
sql = 'select * from results'
```

```python
con = sqlite3.connect('../players.db')
```

```python
df = pd.read_sql(sql,con)
```

```python
players = pd.read_sql('select * from players',con)
```

```python
players
```

```python
am_players = list(players[players['Team']=='America']['Name'])
eu_players = list(players[players['Team']=='Europe']['Name'])
```

```python
ll=[]
```

```python
for player in am_players:
    team='America'
    year = df[df['American_player'].str.contains(player)].iloc[0]['Year']
    temp = df[df['American_player'].str.contains(player)]
    frame = temp[temp['Year']==year]
    losses=frame['Europe_won'].sum()
    total=len(frame)
    wins = total-losses
    pct = round(wins/total*100)
    l = [player,team,year,total,wins,losses,pct]
    ll.append(l)
```

```python
for player in eu_players:
    team='Europe'
    year = df[df['European_player'].str.contains(player)].iloc[0]['Year']
    temp = df[df['European_player'].str.contains(player)]
    frame = temp[temp['Year']==year]
    wins=frame['Europe_won'].sum()
    total=len(frame)
    losses = total-wins
    pct = round(wins/total*100)
    l = [player,team,year,total,wins,losses,pct]
    ll.append(l)
```

```python
fr = pd.DataFrame(data=ll,columns=['Name','Team','Year','Matches','Wins','Losses','Pct'])
```

```python
fr
```

```python
fr.to_sql('rookies',con)
```

```python
con.close()
```
