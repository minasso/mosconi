```python
import pandas as pd
import sqlite3
```

```python
sql='select * from results'
con = sqlite3.connect('../players.db')
df = pd.read_sql(sql,con)
```

```python
#start with easy years that have full team entries already
lst=[]
years=range(2006,2017)
for year in years:
    lst.append(max(set(df[df['Year']==year]['American_player']),key=len))
ser = pd.Series(lst,index=years)
am_roster = ser.to_frame('American_team')
```

```python
am_roster['player1'] = am_roster['American_team'].str.split(',').str.get(0)
am_roster['player2'] = am_roster['American_team'].str.split(',').str.get(1)
am_roster['player3'] = am_roster['American_team'].str.split(',').str.get(2)
am_roster['player4'] = am_roster['American_team'].str.split(',').str.get(3)
am_roster['player5'] = am_roster['American_team'].str.split(',').str.get(4)
```

```python
lst=[]
years=range(2006,2017)
for year in years:
    lst.append(max(set(df[df['Year']==year]['European_player']),key=len))
ser = pd.Series(lst,index=years)
eu_roster = ser.to_frame('European_team')
```

```python
eu_roster['player1'] = eu_roster['European_team'].str.split(',').str.get(0)
eu_roster['player2'] = eu_roster['European_team'].str.split(',').str.get(1)
eu_roster['player3'] = eu_roster['European_team'].str.split(',').str.get(2)
eu_roster['player4'] = eu_roster['European_team'].str.split(',').str.get(3)
eu_roster['player5'] = eu_roster['European_team'].str.split(',').str.get(4)
```

```python
# for years which don't have team matches, use singles matches to create player set
sing = df[df['Format']=='Singles']

rost = []
years = range(1994,2006)
for year in years:
    rost.append(','.join(list(set(sing[sing['Year']==year]['European_player']))).replace(' ,',',').replace(',',', ').strip())

eu_rost = pd.Series(rost,index=years).to_frame('European_team')

eu_rost['player1'] = eu_rost['European_team'].str.split(',').str.get(0)
eu_rost['player2'] = eu_rost['European_team'].str.split(',').str.get(1)
eu_rost['player3'] = eu_rost['European_team'].str.split(',').str.get(2)
eu_rost['player4'] = eu_rost['European_team'].str.split(',').str.get(3)
eu_rost['player5'] = eu_rost['European_team'].str.split(',').str.get(4)
```
