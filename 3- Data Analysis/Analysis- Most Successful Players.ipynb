Goal: Do some analysis on player_stats data to determine most successful players in Mosconi Cup history for various categories

```python
import pandas as pd
import pickle
import os
import sqlite3
```

```python
parent = os.path.abspath('..')
path = os.path.join(parent,'players.db')
```

```python
con = sqlite3.connect(path)
```

```python
with con:
    player_stats = pd.read_sql('select * from playerstats',con)
```

```python
player_stats.head()
```

# Data Analysis

```python
#Most singles wins
sing = player_stats[['Player','Swl','Swin']].sort_values('Swin',ascending=False).head(11).reset_index(drop=True)

#Best singles win percentage
singpct = player_stats[player_stats['Swin']>5][['Player','Swin','Swl','Spct']].sort_values(['Spct','Swin'],ascending=[False,False]).head(10).reset_index(drop=True)

#Most doubles wins
dub = player_stats[['Player','Dwl','Dwin']].sort_values('Dwin',ascending=False).reset_index(drop=True).head(10)

#Best doubles win percentage
dubpct = player_stats[player_stats['Dwin']>5][['Player','Dwl','Dwin','Dpct']].sort_values('Dpct',ascending=False).reset_index(drop=True).head(10)

#Most combined wins
comb = player_stats[['Player','Cwl','Cwin']].sort_values('Cwin',ascending=False).reset_index(drop=True).head(10)

#Best combined win percentage
combpct = player_stats[player_stats['Cwin']>5][['Player','Cwl','Cpct']].sort_values('Cpct',ascending=False).head(11).reset_index(drop=True)
```
