```python
import pandas as pd
```

```python
import sqlite3
```

```python
sql = 'select * from results'
```

```python
con = sqlite3.connect('players.db')
```

```python
df = pd.read_sql(sql,con)
```

```python
df = df.drop('index',axis=1)
```

```python
df
```

```python
f = pd.read_excel('C:\\Users\\rucku\\Desktop\\Coding Projects\\mosconi\\2020-2021 Mosconi cup results.xlsx',sheet_name=1)
```

```python
f['Format'] = f['Europe'].str.extract(r'([A-Z][a-z]+)')
```

```python
f.head()
```

```python
f['European_player'] = f['Europe'].str.extract(r'[A-Z][a-z]+\n?([A-Z].+)')
```

```python
f['American_player'] = f['America'].str.extract(r'[A-Z][a-z]+\n?([A-Z].+)')
```

```python
f
```

```python
n = f['Scores'].str.extract(r'(\d)–(\d)')
```

```python
n.columns=['European_score','American_score']
```

```python
n.at[14,'European_score']=5
```

```python
n.at[14,'American_score']=4
```

```python
f = pd.concat([f,n],axis=1)
```

```python
f.drop('Scores',axis=1,inplace=True)
```

```python
f['Year']= '2021'
```

```python
f = f[['Format', 'European_player', 'American_player',
       'European_score', 'American_score', 'Year']]
```

```python
f
```

```python
f['Match'] = range(1, 18, 1)
```

```python
f
```

```python
f['Europe_won'] = f['European_score']>f['American_score']
```

```python
f['Europe_won'] = f['Europe_won'].apply(lambda x: 1 if x == True else 0)
```

```python
f = f[['Year', 'Match', 'Format', 'European_player', 'European_score',
       'American_score', 'American_player', 'Europe_won']]
```

```python
f.head(2)
```

```python
df.tail(3)
```

```python
new = pd.concat([df,f])
```

```python
new = new.reset_index()
```

```python
new = new.drop('index',axis=1)
```

```python
new
```

```python
new.to_sql('results', con, if_exists = 'replace')
```
