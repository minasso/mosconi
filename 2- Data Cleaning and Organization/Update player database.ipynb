```python
# Get rid of numbers at end of locations
f[f.Location.str.contains('[\d]')]
```

```python
# regex to remove any number or bracket
f.Location = f.Location.str.replace('\d','')
f.Location = f.Location.str.replace('[','')
f.Location = f.Location.str.replace(']','')
```
