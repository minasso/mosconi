```python
f = pd.read_excel('C:\\Users\\rucku\\Desktop\\Coding Projects\\mosconi\\2020-2021 Mosconi cup results.xlsx')
```

```python
f['Format'] = f['Europe'].str.split('\n').str.get(0)
```

```python
f
```

```python
f['Europe'] = f['Europe'].str.replace('Teams\n','').str.replace('Doubles\n','').str.replace('Singles\n','').str.replace('\n',', ')
```

```python
f['America'] = f['America'].str.replace('Teams\n','').str.replace('Doubles\n','').str.replace('Singles\n','').str.replace('\n',', ')
```

```python
f['European_score'] = f['Scores'].str.split('–').str.get(0)
```

```python
f['American_score'] = f['Scores'].str.split('–').str.get(1)
```

```python
f.drop('Scores',axis=1,inplace=True)
```

```python
f.columns=['European_player','American_player','Format','European_score', 'American_score']
```

```python
f['Year']= '2020'
```

```python
f
```

```python
f['Match'] = range(1, 15, 1)
```

```python
f['Europe_won'] = f['European_score']>f['American_score']
```

```python
f
```

```python
f['Europe_won'] = f['Europe_won'].apply(lambda x: 1 if x == True else 0)
```

```python
f = f[['Year', 'Match', 'Format', 'European_player', 'European_score',
       'American_score', 'American_player', 'Europe_won']]
```
