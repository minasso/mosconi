```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/2017_Mosconi_Cup'
res = requests.get(url)
soup = BeautifulSoup(res.text,'lxml')
tables = soup.find_all('table')
tables = tables[4:-1]
```

```python
table=tables[0]
```

```python
rows = table.find_all('tr')[1:-2]
form=[]
am_player=[]
am_score=[]
eu_score=[]
eu_player=[]
for row in rows:
    tds = row.find_all('td')
    i=0
    for td in tds:
        x=td.text.split('\n')
        i+=1
        if i==1:
            form.append(x[0])
            am_player.append(x[1:])
        elif i==2:
            a,b=x[0].split('â€“')
            am_score.append(a)
            eu_score.append(b)
        elif i==3:
            eu_player.append(x[1:])
df = pd.DataFrame({'Format':form,'American_player':am_player[:],
           'American_score':am_score,'European_score':eu_score,
           'European_player':eu_player[:]})
```

```python
df
```

```python
# I had already did a trial run to get the first table's info into an initial dataframe, now i'm looping the logic
for table in tables[1:]:
    rows = table.find_all('tr')[1:-2]
    form=[]
    am_player=[]
    am_score=[]
    eu_score=[]
    eu_player=[]
    for row in rows:
        tds = row.find_all('td')
        i=0
        for td in tds:
            x=td.text.split('\n')
            i+=1
            if i==1:
                form.append(x[0])
                am_player.append(x[1:])
            elif i==2:
                a,b=x[0].split('â€“')
                am_score.append(a)
                eu_score.append(b)
            elif i==3:
                eu_player.append(x[1:])
    df1 = pd.DataFrame({'Format':form,'American_player':am_player[:],
               'American_score':am_score,'European_score':eu_score,
               'European_player':eu_player[:]})
    df=pd.concat([df,df1])
```
