Copy and pasted locations from https://en.wikipedia.org/wiki/Mosconi_Cup and saved in 'locations.csv'

```python
import pandas as pd
```

```python
path = 'locations.csv'
locs = pd.read_csv(path)
```

```python
locs.set_index('Year',inplace=True)
```

```python
d = pd.DataFrame.to_dict(locs)
```

```python
d
```

```python
#get rid of outer dictionary
dic={}
for i in range(1994,2017):
#     print(d['Venue'][i])
    dic[i]=d['Venue'][i]
dic
```

```python
#string replace to get rid of \xa0
dloc={}
for key,val in dic.items():
    val = val.replace(u'\xa0', u' ')
    dloc[key]=val
dloc
```

```python
import pickle
path = 'dloc'
pickle.dump( dloc, open( "dloc", "wb" ) )
```

```python
dloc[1994]
```

```python
df = pickle.load(open('pkl/allyears_with_teams','rb'))
```

```python
df.index.get_level_values('Year')
```

```python
df['Year']= df.index.get_level_values('Year')
df['Location'] = df['Year'].apply(lambda x: dloc[x])
```

```python
df
```

```python
df.drop('Year', axis=1, inplace=True)
```

```python
df
```

```python
df.set_index('Location', append=True, inplace=True).sort_index()
```

```python
df = df.reorder_levels([0,2,1])
```
