```python
import pandas as pd
```

```python
file = 'mosconi_2019.csv'
```

```python
df = pd.read_csv(file, header = None)
```

```python
f = df.fillna(method = 'ffill',axis=0)
```

```python
import numpy as np
```

```python
formats = ['Teams','Doubles','Singles']
```

```python
def fn(x):
    if x in (formats):
        return x
    else:
        return np.nan
```

```python
f['format'] = f[0].apply(fn)
```

```python
f = f.fillna(method = 'ffill',axis=0)
```

```python
f
```

```python
def f2(x):
    if x in formats:
        return np.nan
    else:
        return x
```

```python
f[3]=f[0].apply(f2)
```

```python
f = f.dropna(axis=0)
```

```python
f.drop(3,axis=1,inplace=True)
```

```python
f[1].str.split('–').str.get(0)
```

```python
f['american_score']=f[1].str.split('–').str.get(0)
f['european_score']=f[1].str.split('–').str.get(1)
```

```python
f.drop(1,axis=1,inplace=True)
```

```python
f = f[['format',0, 'american_score', 2,'european_score']]
```

```python
f = f.rename({0:'american_player',2:'european_player'},axis=1)
```

```python
f = f[['format', 'american_player', 'american_score','european_score', 'european_player']]
```

```python
f = f.reset_index(drop=True)
```

```python
f.columns = map(lambda x: x.capitalize(),f.columns)
```

```python
f = f[['Format','European_player','European_score','American_score', 'American_player']]
```

```python
f['Year']='2019'
```

```python
f = f[['Year','Format','European_player','European_score','American_score', 'American_player']]
```

```python
f['Match'] = f.index
```

```python
f = f[['Year','Match','Format','European_player','European_score','American_score', 'American_player']]
```

```python
def f3(x):
    if x.loc['European_score']>x.loc['American_score']:
        return 1
    else:
        return 0
```

```python
f.head()
```

```python
f['Europe_won'] = f.apply(f3,axis=1)
```

```python
f
```

```python
dub = f[f.Format=='Doubles']
other = f[f.Format!='Doubles']

dub1 = dub.iloc[::2]

dub2 = dub.iloc[1::2]

dub1 = dub1.reset_index(drop=True)
dub2 = dub2.reset_index(drop=True)
```

```python
dub1
```

```python
dub2
```

```python
d = dub1.join(dub2,rsuffix='_1')
```

```python
d
```

```python
d = pd.concat([dub1,dub2],axis=1)
```

```python
d['European_player'] = d['European_player']+', '+d['European_player_1']
```

```python
d['American_player'] = d['American_player']+', '+d['American_player_1']
```

```python
d = d[['Year', 'Match', 'Format', 'European_player', 'European_score',
       'American_score', 'American_player', 'Europe_won']]
```

```python
d
```

```python
new = pd.concat([d,other]).sort_values('Match')

new = new.reset_index(drop=True)
```

```python
new['Match']=new.index
```
